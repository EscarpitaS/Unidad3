```python
# EJERCICIO 1. SEBASTIÁN ESCÁRPITA SÁNCHEZ. DISTRIBUCIÓN BINOMIAL
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de parámetros
n_pedidos = 20  # número de pedidos por día
p_exito = 0.70  # probabilidad de éxito por pedido
n_dias = 7  # número de días a simular

# Simulación de pedidos satisfechos por día
np.random.seed(42)  # para reproducibilidad
pedidos_satisfechos = np.random.binomial(n=n_pedidos, p=p_exito, size=n_dias)

# Imprimir resultados
print("Pedidos satisfechos por día:")
for dia, pedidos in enumerate(pedidos_satisfechos, 1):
    print(f"Día {dia}: {pedidos} pedidos satisfechos")

# Cálculo de estadísticos
print("\nEstadísticos de la simulación:")
print(f"Media de pedidos satisfechos: {np.mean(pedidos_satisfechos):.2f}")
print(f"Desviación estándar: {np.std(pedidos_satisfechos):.2f}")
print(f"Mínimo de pedidos satisfechos: {np.min(pedidos_satisfechos)}")
print(f"Máximo de pedidos satisfechos: {np.max(pedidos_satisfechos)}")

# Visualización de la distribución
plt.figure(figsize=(10, 6))
sns.histplot(pedidos_satisfechos, kde=True, stat="density")
plt.title("Distribución de Pedidos Satisfechos")
plt.xlabel("Número de Pedidos Satisfechos")
plt.ylabel("Densidad")
plt.show()

# Cálculo de probabilidades específicas
prob_menos_15 = np.sum(pedidos_satisfechos < 15) / n_dias * 100
prob_mas_18 = np.sum(pedidos_satisfechos > 18) / n_dias * 100

print(f"\nProbabilidad de días con menos de 15 pedidos satisfechos: {prob_menos_15:.2f}%")
print(f"Probabilidad de días con más de 18 pedidos satisfechos: {prob_mas_18:.2f}%")
```

Cada día, el distribuidor tiene una probabilidad de éxito de cumplir un pedido (es decir, de tener en stock los artículos solicitados) del 70%. Desea simular la cantidad de pedidos diarios que logra satisfacer para gestionar adecuadamente sus inventarios y minimizar las faltantes de stock.
