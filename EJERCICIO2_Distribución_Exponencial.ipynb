```python
# EJERCICIO 2. DISTRIBUCIÓN EXPONENCIAL. SEBASTIÁN ESCÁRPITA SÁNCHEZ

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
lambda_exp = 1 / 15  # Tasa de llegada (1 cada 15 minutos)
horas_operacion = 12  # Duración del día en horas
minutos_operacion = horas_operacion * 60  # Convertir a minutos

# Generar tiempos entre llegadas con distribución exponencial
tiempos_entre_llegadas = np.random.exponential(scale=1/lambda_exp, size=1000)

# Filtrar tiempos acumulativos dentro del horario de operación
tiempos_acumulados = np.cumsum(tiempos_entre_llegadas)
tiempos_acumulados = tiempos_acumulados[tiempos_acumulados <= minutos_operacion]

# Mostrar resultados
print(f"Número total de llegadas simuladas en {horas_operacion} horas: {len(tiempos_acumulados)}")
print("Tiempos entre llegadas (minutos):", tiempos_entre_llegadas[:10])

# Graficar histograma de tiempos entre llegadas
plt.hist(tiempos_entre_llegadas, bins=30, edgecolor='black', density=True)
plt.title("Distribución de tiempos entre llegadas de vehículos eléctricos")
plt.xlabel("Tiempo entre llegadas (min)")
plt.ylabel("Frecuencia relativa")
plt.show()
```

Problema 2: Distribución Exponencial
Contexto: Una estación de carga rápida para vehículos eléctricos quiere
simular el tiempo entre la llegada de los vehículos para prever la cantidad
de puntos de carga necesarios durante el día. La llegada de los vehículos
sigue una distribución exponencial con una tasa de 1 llegada cada 15
minutos, en promedio.
Objetivo: Generar tiempos entre llegadas de vehículos eléctricos durante
un día de operación (12 horas)
