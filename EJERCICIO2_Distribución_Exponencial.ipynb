```python
# EJERCICIO 2. DISTRIBUCIÃN EXPONENCIAL. SEBASTIÃN ESCÃRPITA SÃNCHEZ

import numpy as np
import matplotlib.pyplot as plt

# ParÃ¡metros de la simulaciÃ³n
lambda_exp = 1 / 15  # Tasa de llegada (1 cada 15 minutos)
horas_operacion = 12  # DuraciÃ³n del dÃ­a en horas
minutos_operacion = horas_operacion * 60  # Convertir a minutos

# Generar tiempos entre llegadas con distribuciÃ³n exponencial
tiempos_entre_llegadas = np.random.exponential(scale=1/lambda_exp, size=1000)

# Filtrar tiempos acumulativos dentro del horario de operaciÃ³n
tiempos_acumulados = np.cumsum(tiempos_entre_llegadas)
tiempos_acumulados = tiempos_acumulados[tiempos_acumulados <= minutos_operacion]

# Mostrar resultados
print(f"NÃºmero total de llegadas simuladas en {horas_operacion} horas: {len(tiempos_acumulados)}")
print("Tiempos entre llegadas (minutos):", tiempos_entre_llegadas[:10])

# Graficar histograma de tiempos entre llegadas
plt.hist(tiempos_entre_llegadas, bins=30, edgecolor='black', density=True)
plt.title("DistribuciÃ³n de tiempos entre llegadas de vehÃ­culos elÃ©ctricos")
plt.xlabel("Tiempo entre llegadas (min)")
plt.ylabel("Frecuencia relativa")
plt.show()
```
